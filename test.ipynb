```python
from torch_geometric.datasets import QM9
DATA_DIR = "./data/qm9"
dataset = QM9(root=DATA_DIR)
```

```python
d = dataset[100]
print(
    f"x (Node Features):\n{d.x}\n\n"
    f"edge_index (Edge Indices):\n{d.edge_index}\n\n"
    f"edge_attr (Edge Attributes):\n{d.edge_attr}\n\n"
    f"y (Target Values):\n{d.y}\n\n"
    f"pos (Node Positions):\n{d.pos}\n\n"
    f"z (Atomic Numbers):\n{d.z}\n\n"
    f"smiles (SMILES String):\n{d.smiles}\n\n"
    f"name (Molecule Name):\n{d.name}"
)
```

```python
smile_example = Chem.MolFromSmiles(d.smiles)
smile_example
```

```python
mol = dataset[:20].smiles
mmm = []
for m in mol:
    mmm.append(Chem.MolFromSmiles(m))

Draw.MolsToGridImage(mmm, molsPerRow=5, useSVG=True, legends=mol)
```
